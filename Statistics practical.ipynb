# Sampling techniques

```python
df=sns.load_dataset("iris")
df
```

__Simple Random Sampling__

```python
n=10
sample_df=df.sample(n,random_state=22)#through this we fixed it,means whenever we call it there is always same sample of data.
print("Sample random sampling")
sample_df
```

# Systematic sampling
__It follow certain rule__

```python
n=15
new_df=df.iloc[::n]
new_df
```

# Stratified sampling
- Divide the data into groups  based on specific column then select randomly from it

```python
df
```

```python
df["species"].unique()
```

```python
df["species"].value_counts()
```

```python
stratified=df.groupby("species").apply(lambda x:x.sample(min(len(x),5)))
stratified
```

```python
stratified=df.groupby("species", group_keys=False).apply(lambda x:x.sample(min(len(x),3)))
stratified
```

# Cluster Sampling

```python
clusters= df["species"].unique()
clusters
```

```python
df=pd.concat([df[df["species"]==i].sample(2) for i in np.random.choice(clusters,size=2,replace=False)]
```
