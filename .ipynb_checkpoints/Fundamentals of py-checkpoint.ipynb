```python
# nested Condition
n=int(input("Enter a number"))
if n>10:
    if n<=20:
        print(n,"Greater than 10 but less than or equal to  20")
    elif n>20 and n<=50:
        print(n,"Greater than 20 but less than or equal to 50")
    else:
        print(n,"Greater than 50")
else:
    print("Less than 10 or equal to 10")
```

```python
#Passward Authentication
orig_passward="Satish@123"
attempt=0
while attempt<3:
      passward=input("Enter your passward") 
      if  passward==orig_passward:
          print("You can acess")
          break
      else:
          print("Incorrect passward")
          attempt +=1
else:
    print("Account is locked, Try again after 24 hrs") 
      
```

__Control Statements__

```python
#Break
for i in fruits:
    if i=="banana":
        print("Yaa I find it")
        break
    else:
        print("Sorry we can't find it")
```

```python
#Continue
for i in fruits:
    if i=="banana":
        continue
        print("Yaa I got it")
        
    else:
        print("Sorry")
```

```python
for i in fruits:
    if i!="banana":
        continue
        print("Yaa I got it")
        
    else:
        print("Sorry")
```

__*args() and **kwargs()__

```python
def mul(a,b):
        return a*b
def mul_(*args):
    return mul(*args)
mul_(2,3)
```

```python
def emp_info(**kwargs):
    for keys,values in kwargs.items():
        print(f'{keys}: {values}')
emp_info(Name="Satish ",Age=19,Role="Data Scientist")
```

__Exception Handling__

```python
n=int(input("Enter a number: "))
result=100/n
print(result)
```

```python
n=int(input("Enter a number: "))
result=100/n
print(result)
```

```python
try:
    n=int(input("Enter a number: "))
    result=100/n
    print(result)
except:
    print("Please enter a valid number")
```

```python
try:
    n=int(input("Enter a number: "))
    result=100/n
    print(result)
except Exception as e:
    print(e)
```

```python
try:
    n=int(input("Enter a number: "))
    result=100/n
    print(result)
except:
    pass
```

```python
try:
    with open("abc.txt",mode="w") as f:
        f.write("Satish is a good boy.")
except Exception as e:
    print(e)
```

```python
try:
    with open("abc.txt",mode="r") as f:
        print(f.read())
except Exception as e:
    print(e)
```
