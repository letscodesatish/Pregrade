# KDE(Kernel Density Estimate)
In Seaborn, kde=True adds a KDE (Kernel Density Estimate) curve to a plot. It is used to show the smooth probability distribution of continuous data — like a smoothed version of a histogram.

```python
sns.kdeplot(df)
plt.show()
```

```python
sns.kdeplot(df,fill=True)
plt.show()
```

```python
sns.kdeplot(df["sepal_length"],color="g",fill=True,label=f"skewness: {round(df.sepal_length.skew(),2)}")
plt.legend()
plt.show()
```

```python
df
```

```python
sns.kdeplot(data=df,x="sepal_length",hue="species",fill=True,alpha=0.2)
plt.show()
```

```python
sns.kdeplot(data=df,x="petal_width",hue="species",fill=True,alpha=0.2)
plt.show()
```

# Heatmap

```python
df
```

```python
df.drop(columns=["species"],inplace=True)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.yticks(rotation=0)
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.yticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(df.corr(),annot=True,cmap="viridis")
plt.yticks(rotation=0)
plt.show()
```
