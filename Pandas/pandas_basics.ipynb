__Data manupulation in pandas__

```python
#acess any column
import pandas as pd
data={
    "name":["satish","kushal","arjun","aakshant","aditya"],
    "age":[19,21,23,42,41],
    "salary":[20000,25000,21000,41000,80000]
}
df=pd.DataFrame(data)
print(df)
df["bonus"]=df["salary"]*0.1
print(df)
df.insert(0,"emp id",[11,22,33,44,55])
print(df)
```

```python
# update data
df.loc[0,"salary"]=100000
print(df)
```

```python
# adding bonus in salary
df["salary"]=df["salary"]+df["bonus"]
print(df)
```

```python
# removing column
print(df.drop(columns=["bonus"]))#It contain inplace=false by default
print(df)
df.drop(columns=["bonus"],inplace=True)
print(df)
#removing multiple column
df.drop(columns=["age","salary"],inplace=True)
print(df)
```

```python
#removing missing values
print(rf)
rf.dropna(axis=0,inplace=True)
print(rf)
```

```python
# filling missing values
print(rf)
rf.fillna(0,inplace=True)
print(rf)
```

```python
import pandas as pd
data={
    "time":[1,2,3,4,5],
    "values":[10,None,30,40,None]
}
df=pd.DataFrame(data)
print("Before interpolation")
print(df)
df["values"]=df["values"].interpolate(method="linear")
print("After interpolation")
print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
data={
    "x":np.arange(10),
    "y":[1, np.nan, np.nan, 8, 27, np.nan, 125, 216, np.nan, 512]
}
df=pd.DataFrame(data)
print(df)
# interpolate using a 2nd degree polynomial
df["poly_interp"]=df["y"].interpolate(method="polynomial",order=2)
print(df)
# plot original and interpolated data
plt.figure(figsize=(3,3))
plt.plot(df["x"],df["y"],"ro",label="original data")
plt.plot(df["x"],df["poly_interp"],"b.-",label="polynomial interpolation")
plt.legend()
plt.title("polynomial interpolation example")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

__Sorting and Aggregation__

```python
emp={
    "name":["satish","satish","arjun"],
    "salary":[10000,25000,7000]
}
df=pd.DataFrame(emp)
df.sort_values(by="name",ascending=True, inplace=True)
print(df)
# sorting multiple column
df.sort_values(by=["name","salary"],ascending=[True,True],inplace=True)
print(df)
```

```python
#aggregation
print(df["salary"].sum())
print(df["salary"].min())
print(df["salary"].max())
print(df["salary"].mean())
```

```python
# grouping data and then aggregation
emp={
    "name":["satish","aakshant","arjun","prashant","arun"],
    "age":[22,60,43,22,43],
    "salary":[10000,80000,25000,50000,7000]
}
df=pd.DataFrame(emp)
grouped_data=df.groupby("age")["salary"].sum()
print(grouped_data)
grouped_data1=df.groupby(["age","name"])["salary"].sum()
print(grouped_data1)
```

```python
df_customer=pd.DataFrame({
    "id":[1,2,3],
    "name":["satish","kushal","arjun"]
})
df_order=pd.DataFrame({
    "id":[1,2,4],
    "order_amount":[200,100,300]
})
df_merge=pd.merge(df_customer,df_order,on="id",how="inner")
print("inner join")
print(df_merge)
df_merge1=pd.merge(df_customer,df_order,on="id",how="outer")
print("outer join")
print(df_merge1)
df_merge2=pd.merge(df_customer,df_order,on="id",how="left")
print("left join")
print(df_merge2)
df_merge3=pd.merge(df_customer,df_order,on="id",how="right")
print("right join")
print(df_merge3)
df_merge4=pd.merge(df_customer,df_order,how="cross")
print("cross join")
print(df_merge4)
```
