```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Data structure :[restaurant_id 2021,2022,2023,2024]
sales_data=np.array([
    [1,150000,180000,220000,250000], #Paradise Biryani
    [2,120000,140000,160000,190000], #Beijing Bites
    [3,200000,230000,260000,300000], #Pizza Hub
    [4,180000,210000,240000,270000], #Burger Points
    [5,160000,185000,205000,230000] #Chai Point
])
print("=== Zomato sales analysis === ")
print("Sales data shape",sales_data.shape)
print("Sample of first 3 restaurant:\n",sales_data[:3])

    
```

```python
# total sales per year
print(np.sum(sales_data,axis=0))
yearly_total=np.sum(sales_data[:,1:],axis=0)
print("yearly total: ",yearly_total)

# minimum sales per restaurent
min_sales=np.min(sales_data[:,1:],axis=1)
print("Min sales per years: ",min_sales)
```

```python
# maximum sales per year
max_sales=np.max(sales_data[:,1:],axis=0)
print(max_sales)
```

```python
# average sales per restaurant
average=np.mean(sales_data[:,1:],axis=1)
print(average)
```

```python
#Cumulative sales
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

```python
#Graph ploting
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales accross all restaurant")
plt.xlabel("Years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

__vector operations__

```python
import numpy as np
vec1=np.array([1,2,3,5])
vec2=np.array([6,3,4,8])
print("Vector addition: ",vec1+vec2)
print("Vector multiplication: ",vec1*vec2)
print("Dot product: ",np.dot(vec1,vec2))
angle=np.arccos(np.dot(vec1,vec2)/(np.linalg.norm(vec1)*np.linalg.norm(vec2)))
print(angle)
```
