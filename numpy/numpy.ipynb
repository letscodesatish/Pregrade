__Array reshaping__

```python
original=np.arange(12)
print("original array: ",original)
reshape=original.reshape(3,4)
print("Reshape array: ",reshape)
flatten=reshape.flatten()
print("flatten array: ",flatten)
```

```python
#ravel (return views ,instead of copy)
revealed=reshape.ravel()
print("ravel array: ",revealed)
```

__Transpose__

```python
print(reshape)
transpose=reshape.transpose()
print("Transpose:\n ",transpose)
```

__sorting__

```python
arr_2d_unsorted = np.array([[111,2],[23,56],[78,43]])
print(arr_2d_unsorted)
print("sorted 2D array: \n",np.sort(arr_2d_unsorted, axis=1))
```

__Filter__

```python
numbers=np.arange(1,11)
print(numbers)
even_num=numbers[numbers %2==0]
print(even_num)
```

__Filter with mask__

```python
sk=numbers>5
print("Numbers greater than 5: ",numbers[sk])
```

__Fancy indexing vs np.where()__

```python
indices=[0,1,2]
print(numbers[indices])
where_result=np.where(numbers>5)
print("NP where",numbers[where_result])
```

```python
condition_array=np.where(numbers>5,numbers*2,numbers)
print(condition_array)
#just like if else stm
```

__Addition and removal of array__

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
arr2=np.array([6,7,8,1,10])
combine=arr1+arr2
print(combine)
concatenate=np.concatenate([arr1,arr2])
print(concatenate)
```

```python
a=np.array([1,2])
b=np.array([9,12,99])
print(a.shape == b.shape)
```

```python
original=np.array([[1,2],[3,4]])
new_row=np.array([[44,21]])
with_new_row=np.vstack((original,new_row))
print("row add \n",with_new_row)
new_col=np.array([[5],[8]])
with_new_col=np.hstack((original,new_col))
print("column add \n",with_new_col)
```

```python
a=np.array([1,2,3,4])
delete=np.delete(a,2)
print("Array after deletion: ",delete)
b=np.array([[1,2],[34,78]])
delete_2d_array=np.delete(b,[0,0])
print("2d_Array after deletion: ",delete_2d_array) 
```
