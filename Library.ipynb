```python
df1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
df2=pd.DataFrame({"C":[11,12,13,14,15],"D":[101,202,303,404,505]})
```

```python
print(df1)
print(df2)
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
def INR_to_Dollar(Salary):
    dollar=[]
    for i in Salary:
        d=i/85.50
        dollar.append(f"{round(d,2)} USD")
    return dollar
```
